```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import re
import os
```

```python
opti = webdriver.ChromeOptions()
opti.add_argument('--headless')
#opti.to_capabilities()
```

```python
li = []
count = 0;
while ( count < 26 ):    
    
    driver = webdriver.Chrome(options=opti)
    driver.get('https://www.medindia.net/drug-price/index.asp?alpha=' + chr(ord('A') + count))
    time.sleep(3)
    
    count += 1
    soup1 = BeautifulSoup(driver.page_source, 'html5lib')
    da = soup1.find('table', class_ = 'table-bordered').find_all('td')
    for i in da:
        li.append(i.text)
    print(driver.current_url)
    
```

```python
remove_ = []
count = 0
for i in li:
    if i.replace('\n',' ').replace(' ','') == 'ViewPrice':
        remove_.append(count)
    if i.isnumeric():
        remove_.append(count)
    if i == '-':
        remove_.append(count)
    count += 1
```

```python
li1 = []
ptr = 0
for i in range(5,len(li)-1):
    if remove_[ptr] == i:
        ptr += 1;
        continue
    else:
        li1.append(li[i])
```
